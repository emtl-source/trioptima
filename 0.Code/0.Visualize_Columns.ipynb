```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

# Your file path
your_path = 'C:/Users/OskarHansson/trioptima/'
file_name = 'Modified_IRS_All_Train_VF_Cash_Risk'
file_path = your_path + "3.Cash_Risk/" + file_name + '.csv'
data = pd.read_csv(file_path)

# Specify the differnet date formats in the dataset
date_formats = ['%Y-%m-%d', '%d/%m/%Y', '%Y-%m-%d %H:%M', '%d-%b-%y', '%Y-%m-%dT%H:%M:%SZ']

# Set a threshold for making a bar chart, if more unique values than below make a histogram
column_threshold = 50

nanCols = []
boolCols = []
strCols = []
dateCols = []
numCols = []

# Function to check for datetime columns
def has_datetime(unique_values):   
    for value in unique_values:
        for format in date_formats:
            try:
                # Return True if any value is on datetime format
                datetime.strptime(value, format)
                return True
            except:
                continue
    return False


# Function to convert values to datetime
def get_dates(column):
    dates = []
    for value in data[column]:
        for format in date_formats:
            try:
                date = datetime.strptime(value, format)
                dates.append(date)
                break
            except:
                continue
    return dates

# Function to print a histogram - for continous values
def print_hist(column):
    plt.figure(figsize=(8, 6))
    num_bins = 10  # Set the number of bins for the histogram
    n, bins, patches = plt.hist(data[column].dropna(), bins=num_bins, color='blue', edgecolor='black')

    # Set x-axis labels to represent only bin edges
    plt.xticks(bins, rotation=45, ha='right')

    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.tight_layout()  # Automatically adjust subplot params for better layout
    plt.show()
    print('=' * 50)

# Function to print a bar chart - for categorical values
def print_bar_chart(distribution):
    
    # Plot bar chart for the sorted data
    plt.figure(figsize=(8, 6))
    distribution.plot(kind='bar', color='blue', edgecolor='black')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    plt.show()

    print('=' * 50)



# 0. CATEGORIZE COLUMNS
for column in data.columns:
    unique_values = data[column].unique()

    # Check if column is empty
    is_nan = data[column].isna().all()

    # Check if col is of type boolean
    has_boolean = any(isinstance(value, bool) for value in unique_values) or data[column].dtype == np.dtype('bool')

    # Check if col is of type string
    has_str = any(isinstance(value, str) for value in unique_values)


    if is_nan:
        nanCols.append(column)
    elif has_boolean:
        boolCols.append(column)
    elif has_str:
        # Check if string is of type date
        if has_datetime(unique_values):
            dateCols.append(column)
        else:
            strCols.append(column)
    else:
        # Else the column is of type number
        numCols.append(column)


# 2. PRINT EMPTY COLUMNS
print('The follwing columns are empty:')
for column in nanCols:
    print(column)
print('=' * 50)


# 3. PRINT NUMERICAL VALUES
for column in numCols:
    print("Column:", column)
    print("Data type: Int")
    
    # Calculate and print minimum, maximum, average, median
    print("Minimum:", data[column].min())
    print("Maximum:", data[column].max())
    print("Average:", data[column].mean())
    print("Median:", data[column].median())
    
    # Count and print the number of NaN values in the column
    num_nan = data[column].isna().sum()
    percent_nan = (num_nan / len(data[column])) * 100
    print("Percentage of NaN values:", percent_nan, "%")

    num_unique_values = data[column].nunique()
    print('Number of unique values:', num_unique_values)

    # Print value counts (distribution) excluding NaN values
    distribution = data[column].value_counts(dropna=True)
    print("Value Counts:")
    print(distribution)
    
    if num_unique_values > column_threshold:
        print_hist(column)
    else:
        sorted_distribution = distribution.sort_index()
        print_bar_chart(sorted_distribution)


# 2. PRINT DATE VALUES
for column in dateCols:
    print("Column:", column)
    print("Data type: Date")
    
    dates = get_dates(column)

    # Calculate average time
    total_timestamp = sum(date.timestamp() for date in dates)
    average_timestamp = total_timestamp / len(dates)
    average_date = datetime.fromtimestamp(average_timestamp)

    print("Minimum:", min(dates))
    print("Maximum:", max(dates))
    print("Average:", average_date)

    # Count and print the number of NaN values in the column
    num_nan = data[column].isna().sum()
    percent_nan = (num_nan / len(data[column])) * 100
    print("Percentage of NaN values:", percent_nan, "%")

    num_unique_values = data[column].nunique()
    print('Number of unique values:', num_unique_values)

    # Print value counts (distribution) excluding NaN values
    distribution = data[column].value_counts(dropna=True)
    print("Value Counts:")
    print(distribution)

    if num_unique_values > column_threshold:
        plt.figure(figsize=(8, 6))
        num_bins = 10  # Set the number of bins for the histogram
        n, bins, patches = plt.hist(dates, bins=num_bins, color='blue', edgecolor='black')
        plt.xticks(bins, rotation=45, ha='right')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {column}')
        plt.tight_layout()  # Automatically adjust subplot params for better layout
        plt.show()
        print('=' * 50)

    else:
        print_bar_chart(distribution)

# 3. PRINT CATEGORICAL VALUES

for column in strCols:
    print("Column:", column)
    print("Data type: String")

    # Count and print the number of NaN values in the column
    num_nan = data[column].isna().sum()
    percent_nan = (num_nan / len(data[column])) * 100
    print("Percentage of NaN values:", percent_nan, "%")

    num_unique_values = data[column].nunique()
    print('Number of unique values:', num_unique_values)

    # Print value counts (distribution) excluding NaN values
    distribution = data[column].value_counts(dropna=True)
    print("Value Counts:")
    print(distribution)

    # Plot bar chart for the column
    print_bar_chart(distribution)



# 4. PRINT BOOLEAN VALUES
for column in boolCols:
    print("Column:", column)
    print("Data type: Boolean")

    # Count and print the number of NaN values in the column
    num_nan = data[column].isna().sum()
    percent_nan = (num_nan / len(data[column])) * 100
    print("Percentage of NaN values:", percent_nan, "%")

    # Print value counts (distribution) excluding NaN values
    distribution = data[column].value_counts(dropna=True)
    print("Value Counts:")
    print(distribution)

    print_bar_chart(distribution)


```
