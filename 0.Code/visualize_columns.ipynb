```python
import pandas as pd
import matplotlib.pyplot as plt

# Your file path
your_path = 'C:/Users/OskarHansson/trioptima/'
file_name = 'RT.IRS_Data_2023'
file_path = your_path + "1.Original/" + file_name + '.csv'

timeCols = ["disseminationTimestamp", "effectiveDate", "eventDateTime", "executionDateTime", "expirationDate", "leg1NotionalEffectiveDate", "leg2NotionalEffectiveDate", "sDRreceiptTimestamp"]

# Read the specified columns from the CSV file into a DataFrame
data = pd.read_csv(file_path, usecols=timeCols)


# List of date-time formats to parse
date_formats = ['%Y-%m-%d %H:%M', '%d-%b-%y', '%Y-%m-%dT%H:%M:%SZ']

# Convert columns to datetime using the specified formats
for column in data.columns:
    for date_format in date_formats:
        try:
            data[column] = pd.to_datetime(data[column], format=date_format, errors='coerce')
            break
        except (ValueError, TypeError):
            print('Error')


for column in data.columns:
    # Print the number of unique values for each column and create histograms
    num_unique_values = data[column].nunique()
    data_type = data[column].dtype
    distribution = data[column].value_counts(dropna=False)
    print(f"Column: {column}")
    print(f"Number of Unique Values: {num_unique_values}")
    print(f"Distribution (including None values):\n{distribution}\n")
    
    try:
        min_value = data[column].min()
        max_value = data[column].max()
        print(f"Minimum Value: {min_value}")
        print(f"Maximum Value: {max_value}")
    except:
        pass

    # Create histogram with bars and centered labels
    plt.figure(figsize=(8, 6))
    plt.hist(data[column].dropna(), bins=20, color='skyblue', edgecolor='black', align='mid')  # Set align='mid'
    
    # Set x-axis ticks to display centered labels
    plt.xlabel(column, fontsize=12)  # Set x-axis font size to 12
    plt.ylabel('Frequency', fontsize=12)  # Set y-axis font size to 12
    plt.title(f'Histogram of {column}', fontsize=14)  # Set title font size to 14
    plt.show()

    print('-' * 50)



```
