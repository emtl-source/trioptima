```python
import pandas as pd
import matplotlib.pyplot as plt

# Your file path
your_path = 'C:/Users/OskarHansson/OneDrive - STRGM AB/Skola/TriOptima/trioptima/'
file_name = 'RT.IRS_Data_2023'
file_path = your_path + "1.Original/" + file_name + '.csv'

# Set to True to plot histogram for each column
plot_hist = True

# Set a threshold for maximum amount of different values for printing a histogram
col_threshold = 10 

# Read the CSV file into a DataFrame
data = pd.read_csv(file_path, dtype=str)

# Get the unique values and number of unique values for each column
for column in data.columns:
    unique_values = data[column].unique()
    unique_values_count = len(unique_values)
    distribution = data[column].value_counts(dropna=False)

    print(f"Column: {column}")
    print(f"Number of unique values: {unique_values_count}\n")
    print(f"Distribution (including None values):\n{distribution}\n")


    # If the column has 10 or fewer unique values, plot histogram
    if unique_values_count <= col_threshold and plot_hist:
        
        # Plot histogram
        plt.figure(figsize=(8, 6))
        data[column].value_counts(dropna=False).plot(kind='bar', edgecolor='black')
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.show()

    print('-' * 50)

```
